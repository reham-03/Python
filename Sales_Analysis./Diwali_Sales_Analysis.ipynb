<span style="color:#A52A2A"><b>Introduction:</b></span>üìä

This notebook presents an in-depth analysis of the Diwali sales dataset. Diwali, the festival of lights, is one of the most significant festivals in India, and it is marked by extensive shopping and gifting activities. Retailers often witness a substantial increase in sales during this period. The aim of this analysis is to uncover trends, insights, and patterns within the dataset to better understand consumer behavior and sales performance during Diwali.

<span style="color:#A52A2A"><b>Objectives:</b></span>üìä üéØ

1. **Sales Trends Analysis**: Identify overall sales trends during the Diwali season.
2. **Category-wise Performance**: Determine which product categories performed best.
3. **Customer Demographics**: Examine customer demographics to understand purchasing behavior.
4. **Regional Analysis**: Investigate sales distribution across different regions.

Improved sales by identifying most selling product categories and products, which can help to plan inventory and hence meet the demands

# Exploratory Data Analysis (EDA)

```python
ax=sns.countplot(x = 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data=df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
sns.barplot(x = 'Gender',y= 'Amount',data=data)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
ax=sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#ax=sns.countplot(data = df, x = 'Married', hue = 'Gender',palette='Set2')
ax=sns.countplot(data = df, x = 'Married', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount' ,data = sales_age)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
total=sum(sales_age['Amount'])
for i, v in enumerate(sales_age["Amount"]):
    percentage = v/total *100
    plt.text(i, v, f'{percentage:.2f}%', ha="center", va="bottom")

```

# Married

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',palette='Set2')
```

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
